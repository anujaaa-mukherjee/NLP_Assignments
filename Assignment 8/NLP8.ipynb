```python
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
```

```python
sentence = "I went to the bank to deposit money"
target_word = "bank"

print("Sentence:", sentence)
print("Target Word:", target_word)
```

```python
nltk.download('punkt_tab')
sense = lesk(word_tokenize(sentence), target_word)

print("\n--- Disambiguated Sense ---")
print("Sense:", sense)
print("Definition:", sense.definition())
```

```python
print("\n--- All Possible Meanings of 'bank' ---")

for syn in wordnet.synsets(target_word):
    print(syn, ":", syn.definition())
```

```python
sentences = [
    ("He sat on the bank of the river", "bank"),
    ("She deposited cash in the bank", "bank"),
    ("The bat flew in the night", "bat"),
    ("He used a bat to hit the ball", "bat")
]

print("\n--- Multiple Sentence Disambiguation ---")

for sent, word in sentences:
    sense = lesk(word_tokenize(sent), word)
    print(f"\nSentence: {sent}")
    print(f"Word: {word}")

    if sense:
        print("Predicted Meaning:", sense.definition())
    else:
        print("No sense found")
```

```python
def disambiguate(sentence, word):
    sense = lesk(word_tokenize(sentence), word)
    if sense:
        return sense.definition()
    else:
        return "No meaning found"
```

```python
print("\n--- Function Output ---")
print(disambiguate("The crane is flying in the sky", "crane"))
```
