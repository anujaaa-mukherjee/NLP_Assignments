```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import Counter
```

```python
text = """
Natural Language Processing is a field of Artificial Intelligence.
It helps computers understand human language.
"""

print("Original Text:\n", text)
```

```python
import nltk
nltk.download('punkt_tab')
tokens = word_tokenize(text)

print("\n--- Tokens ---")
print(tokens)
```

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("\n--- After Stopword Removal ---")
print(filtered_tokens)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("\n--- Lemmatized Tokens ---")
print(lemmatized_tokens)
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
pos_tags = nltk.pos_tag(lemmatized_tokens)

print("\n--- POS Tags ---")
print(pos_tags)
```

```python
word_freq = Counter(lemmatized_tokens)

print("\n--- Word Frequency ---")
for word, freq in word_freq.items():
    print(f"{word}: {freq}")
```

```python
def nlp_pipeline(text):
    tokens = word_tokenize(text)
    tokens = [w for w in tokens if w.isalpha()]
    tokens = [w for w in tokens if w.lower() not in stopwords.words('english')]
    tokens = [WordNetLemmatizer().lemmatize(w) for w in tokens]
    pos = nltk.pos_tag(tokens)
    freq = Counter(tokens)

    return {
        "tokens": tokens,
        "pos_tags": pos,
        "frequency": freq
    }
```

```python
print("\n--- Using NLP Pipeline Function ---")
result = nlp_pipeline("NLP makes machines intelligent and useful.")

print("Tokens:", result["tokens"])
print("POS Tags:", result["pos_tags"])
print("Frequency:", result["frequency"])
```
