```python
from googletrans import Translator

translator = Translator()
```

```python
text = """
India is one of the fastest growing economies in the world.
The government is focusing on digital transformation and smart cities.
"""

print("Original English Text:\n", text)
```

```python
translated_hi = translator.translate(text, src='en', dest='hi')

print("\nTranslated to Hindi:\n", translated_hi.text)
```

```python
translated_en = translator.translate(translated_hi.text, src='hi', dest='en')

print("\nBack to English:\n", translated_en.text)
```

```python
languages = {
    "Marathi": "mr",
    "Tamil": "ta",
    "Gujarati": "gu"
}

print("\n--- Translation into Multiple Indian Languages ---")

for lang, code in languages.items():
    translated = translator.translate(text, dest=code)
    print(f"\n{lang}:\n{translated.text}")
```

```python
def translate_text(text, target_lang):
    result = translator.translate(text, dest=target_lang)
    return result.text
```

```python
print("\nCustom Function Example (English → Marathi):")
print(translate_text("Education is important for success.", "mr"))
```
